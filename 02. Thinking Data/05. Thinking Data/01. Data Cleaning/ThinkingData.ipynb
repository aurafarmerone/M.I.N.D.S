```python
import json
```

```python
#Load data
def load_data(filename):
    with open(filename,"r") as f:
        data = json.load(f)
    return data    
```

```python
data = load_data("customer_data.json")
print("Data of Customers \n")
print(data)
```

```python
#cleaning data four --> 4
def clean_data(data):
    # creating map to fix numberic data
    Text_To_Num = {"one" : '1' , "two" : '2', "three" : '3' , "four": '4', "five": '5'}
    for user in data:
        rating = user["rating"].strip().lower() # remove spaces, FOUR-> four

        if(rating in Text_To_Num): # four in {"four": 4} ---> True
            rating = Text_To_Num[rating] # putting value from key
        user["rating"] = rating # overide in original data

# def float_to_int(data):
#     # 4.4 to 4
#     for user in data:
#         floatdata = user["rating"]
#         if(0 <= floatdata <= 9):
#             floatdata = (int)user["rating"]
#             user["rating"] = floatdata
            
        
    
def print_data(data):
    for user in data:
        print(user)
        

print("\nBefore Fix")
print_data(data)

clean_data(data)

print("\nAfter Fix")
print_data(data)


# print("After Fix Float value")
# float_to_int(data)
# print_data(data)
```

```python
# handle missing age value
def printAge(data):
    for user in data:
        print(user.get("age"))
    
def handleMissing(data):
    for user in data:
        agedata = user.get("age")
        if(agedata == None):
            user["age"] = "Null"
        

print("##### Before #####")
printAge(data)

handleMissing(data)

print("##### After #####")
printAge(data)
```

```python
def noduplicate(data):
    no_duplicate = []
    unique_data = set()

    for user in data:
        name = user["name"].strip()

        if name in unique_data:
            continue
        else:
            unique_data.add(name)   # â­ missing line
            no_duplicate.append(user)

    print(no_duplicate)

noduplicate(data)
```
